以下のセルで足し算を計算してみましょう。`1+1`は式で、単純なプログラム命令となっていて、Python処理系はこの命令を評価して足し算の結果を返しています。セルの式は`Ctrl+Enter`もしくは`Shift+Enter`で実行することができます。

式の中の`1`は値、`+`は数学演算子です。

```python
1+1
```

数学演算子には他に以下のような演算子があります。これらの演算子を使って式を入力して実行し、その評価結果を確認してみてください。
- `**`： 累乗
-  `*`： 掛け算
- `/`： 割り算
- `//`: 整数の割り算（小数点以下は切り捨て）
- `%`: 割り算の余り
- `+`: 足し算
- `-`: 引き算
